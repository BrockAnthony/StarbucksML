#### Handling Missing Values

```python
transcript.isnull().sum()
```

```python
profile.isnull().sum()
```

There are 2175 missing values in the "gender" column of the profile dataframe, which is a relatively small % of the dataset, so I'll replace the missing values with the column's mode. There are also the same amount of missing values in the "income" column of that dataframe, which I'll fill with the mean of the column's values. 

```python
profile['gender'] = profile['gender'].fillna(profile['gender'].mode()[0])
```

```python
profile['income'] = profile['income'].fillna(profile['income'].mean())
```

```python
profile.isnull().sum()
```

```python
#Splitting channels column
channel_list = []

for channel in portfolio['channels']:
    for i in channel:
        if i not in channel_list:
            channel_list.append(i)
            
for channel in channel_list : 
    portfolio[channel] = portfolio['channels'].apply(lambda x:
                                                     1 if channel in x else 0)
    
portfolio.drop('channels', axis = 1, inplace = True)
```
